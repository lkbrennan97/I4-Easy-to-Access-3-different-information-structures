### Securely Storing API Keys in Colab

To securely store your API keys in Colab, follow these steps:

1.  **Open the Secrets tab:** Click on the 'key' icon (🔑) in the left-hand panel of your Colab notebook.
2.  **Add a new secret:** Click 'Add new secret'.
3.  **Name the secret:** For this notebook, you should name your secret `LASTFM_API_KEY` (or whatever name you prefer, but ensure it matches the name you use in your code).
4.  **Enter the secret value:** Paste your actual Last.fm API key into the 'Value' field.
5.  **Save the secret:** Click 'Save secret'.
6.  **Enable notebook access:** Make sure 'Notebook access' is toggled ON for this secret.

```python
from google.colab import userdata

# Retrieve the API key securely from Colab Secrets
MY_API_KEY = userdata.get('LASTFM_API_KEY')
```

```python
# LastFM API
# https://www.last.fm/api/show/artist.getInfo

import requests

def get_artist_info(artist_name, api_key):
    """
    Fetches artist metadata from the Last.fm API.
    """
    url = "http://ws.audioscrobbler.com/2.0/"

    # Define the parameters for the API call
    params = {
        'method': 'artist.getinfo',
        'artist': artist_name,
        'api_key': api_key,
        'format': 'json'
    }

    try:
        response = requests.get(url, params=params)
        response.raise_for_status()  # Checks for HTTP errors
        return response.json()
    except requests.exceptions.RequestException as e:
        return {"error": str(e)}

# Usage
if __name__ == "__main__":
    # MY_API_KEY is retrieved from google.colab.userdata in a previous cell.
    if not MY_API_KEY:
        print("Error: LASTFM_API_KEY not found. Please ensure it's set in Colab secrets and notebook access is enabled.")
    else:
        data = get_artist_info("Children of Bodom/", MY_API_KEY)

        if "artist" in data:
            artist = data['artist']
            print(f"Artist: {artist['name']}")
            print(f"Listeners: {artist['stats']['listeners']}")
            print(f"Summary: {artist['bio']['summary']}")
        else:
            print("Could not find artist info.")
```
